LOAN APPROVAL PREDICTION USING RANDOM FOREST


```python
import pandas as pd
```

```python
df= pd.read_csv('/content/loan_data (1).csv')
df
```

```python
d=pd.concat([(df.isnull().sum())],axis=1,keys=['df'])
print(d)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
target=df["Loan_Status"]
print(target)
```

```python
feat_list=['Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']
feat=df[feat_list]
print(feat)
```

```python
model=RandomForestClassifier()
model.fit(feat,target)
```

```python
DT=model.fit(feat,target)
```

```python
Dependents=int(input("Enter Dependents: "))
ApplicantIncome=int(input("Enter ApplicantIncome: "))
CoapplicantIncome=int(input("Enter CoapplicantIncome: "))
LoanAmount=int(input("Enter LoanAmount: "))
Loan_Amount_Term=int(input("Enter Loan_Amount_Term: "))
Credit_History=int(input("Enter Credit_History: "))
pred=model.predict([[Dependents,ApplicantIncome,CoapplicantIncome,LoanAmount,Loan_Amount_Term,Credit_History]])
print(pred)
```

```python
if pred == ['Y']:
  print("We are glad to inform that your Loan will be approved")
else:
  print("We are sorry,your Loan will not be approved")
```
