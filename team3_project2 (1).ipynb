LOAN APPROVAL PREDICTION USING DECISION TREE

```python
import pandas as pd
```

```python
df= pd.read_csv('/content/loan_data (1).csv')
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
target=df["Loan_Status"]
print(target)
```

```python
feat_list=['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
feat=df[feat_list]
print(feat)
```

```python
model=DecisionTreeClassifier()
model.fit(feat,target)
```

```python
DT=model.fit(feat,target)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(DT)
```

```python
ApplicantIncome=int(input("Enter ApplicantIncome: "))
CoapplicantIncome=int(input("Enter CoapplicantIncome: "))
LoanAmount=int(input("Enter LoanAmount: "))
Loan_Amount_Term=int(input("Enter Loan_Amount_Term: "))
pred=model.predict([[ApplicantIncome,CoapplicantIncome,LoanAmount,Loan_Amount_Term]])
print(pred)
```

```python
if pred == ['Y']:
  print("We are glad to inform that your Loan will be approved")
else:
  print("We are sorry,your Loan will not be approved")
```
